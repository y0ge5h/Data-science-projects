```python
#Function for getting the top ten similar users for a given user
def get_similarity(user_id):
    similarity = pd.DataFrame(corr_matrix[user_id])
    similarity["username"] = list(corr_matrix)[:-1]
    similarity = similarity.sort_values(by=user_id,ascending=False)
    similarity = similarity[similarity[user_id]<0.99]
    return similarity.head(10)
```

```python
similar_users = get_similarity(4589)
similar_users
```

```python
#Function for getting the challenges solved by the similar users
def get_recommended_users(similar_users):
    target_user = list(similar_users)[0]
    similar_usernames =[]
    for user in similar_users["username"]:
        similar_usernames.append(user)
    df_recommend = df.loc[[target_user,similar_usernames[0],similar_usernames[1],similar_usernames[2]],:]
    return df_recommend
```

```python
rec_users = get_recommended_users(similar_users)
```

```python
#Function for getting the top 3 recommended challenges from the top 10 similar users
def get_recommended_challenges(rec_users):
    recomended_challenges = []
    i=0
    while(i<rec_users.shape[0]):
        for challenge in list(rec_users):
            if rec_users.loc[:,challenge].iloc[0]==0 and rec_users.loc[:,challenge].iloc[i]==1:
                recomended_challenges.append(challenge)
        i+=1
    reco_list = list(set(recomended_challenges))
    return reco_list[:3]
```

```python
recommended_chal={}
test_users = test_df.index.values.tolist()[:10]
for user in test_users:
    similar_users = get_similarity(user)
    rec_users = get_recommended_users(similar_users)
    recommended_chal[user] = get_recommended_challenges(rec_users)
```
